```python
#creation of class node . 
class node:
    def __init__(self,data):
        self.data = data
        self.prev = None
        self.next = None

# creation of double linked list as DDL.        
class DDL:
    def __init__(self):
        self.head = None # making the head has None .
    
    def insertb(self,element): # creation of insertb for adding the element in front of the linked list . 
        new_node =  node(element) # making new node by calling the node class . 
        temp = self.head # creating a temporary variable to store self.head 
        temp.prev = new_node # pointing the temp.ref to the new_node
        new_node.next = temp # making the new_node next has temp means chaning the head . 
        self.head = new_node # pointing the self.head to the new_node . 
    
    def inserte(self,element): # creation of insertb ha 
        new_node = node(element)
        temp = self.head 
        while temp.next is not None:
            temp = temp.next
        temp.next = new_node
    
    def insertm(self,pos,element):
        new_node = node(element)
        temp = self.head.next
        pre = self.head
        for i in range(pos - 1):
            temp = temp.next
            pre = pre.next
        new_node.data = element 
        new_node.prev = pre
        pre.next = new_node
        temp.prev = new_node
        new_node.next = temp
        
    def removeb(self):
        temp = self.head 
        self.head = temp.next
        temp.next = None
        
    def remove_e(self):
        temp = self.head.next
        pre = self.head 
        while temp.next is not None:
            temp = temp.next
            pre = pre.next
        pre.next = None
    
    def removem(self,pos):
        temp = self.head.next 
        pre = self.head
        for i in range(pos - 1):
            temp = temp.next 
            pre = pre.next 
        pre.next = temp.next
        
        
        
        
        
    
    def dis(self):
        if self.head is None:
            print("linked list is empty .")
        else:
            temp = self.head
            while temp:
                print("-->> ",temp.data,end= ' ')
                temp = temp.next
                
                
obj = DDL()
n1 = node(123)
obj.head = n1

obj.inserte(23)
obj.insertb(12)
obj.insertb(1)
obj.inserte([12])
obj.insertb(16)
obj.inserte(["done"])
obj.insertm(2,[12,3,5])
obj.inserte(12)
obj.insertb(145)
obj.insertm(2,[12,3,4,5,6])
obj.removeb()
obj.remove_e()
obj.removeb()
obj.removem(1)
obj.insertb([12,3,45])


obj.dis()
```

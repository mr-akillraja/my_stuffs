```python


#creation node class .
class node:
    def __init__(self,data):
        self.data = data
        self.ref = None

#creation of singly linked list class using the variable "SSL"
class SLL:
    def __init__(self):
        self.head = None
        self.length = 0
        
    def node_begin(self,element):
        new_element = node(element) # creation of node -->> by callling of node class.
        new_element.ref = self.head # linking the head node refernece to the new_node.
        self.head = new_element    #  head is changed to new_node.
        self.length += 1
        
    def node_end(self,element):
        new_element = node(element) # creation of node -->> by calling of node class . 
        temp = self.head # checking the last node whether the pointer is of "NULL" value . 
        while temp.ref: # traversal through the linked list to find the last node . 
            temp = temp.ref # increment value to find the last node . 
        temp.ref = new_element # last is noded is founded,then the new_node becomes the last_node using temp variable . 
        
        
    def node_middle(self,pos,element):
        new_node = node(element) # creation of new node -->> by calling of node class . 
        temp = self.head # creating temporary variable and having value self.head 
        for i in range(pos-1): # looping through the linked list using the index till the index given by using temp.ref
            temp = temp.ref # increment of the loop for going through the loop till the loop index given
        new_node.data = element # adding the new node value to the new_node.data . so that the value is store in data field.
        new_node.ref = temp.ref # making the ref to point to next node ref. so that the ref are interchanged .
        temp.ref = new_node # storing the old node ref to the new_node ref link.
        
        
    def del_beg(self):
        temp = self.head # creating a temporary variable with value self.head pointing to the head node of linked list . 
        self.head = temp.ref # changing the head node by pointing the head node to the second of the linked list .
        temp.ref = None # making the new node reference has "NONE" .

    
    def del_end(self):
        temp = self.head.ref # creating a temporary variable. so that temp will point at next node of head node..means starting node+1
        prev = self.head # creating a prev variable so that prev is pointing to the head node . 
        while temp.ref is not None: # it iterates until the "NONE" occuurs .
            temp = temp.ref # this is for increment for the temporary variable .
            prev = prev.ref # this is for increment for the prev variable . 
        prev.ref = None # removing the last node and making the prev node has "NONE" .
        
        
    def del_pos(self,index):
        temp = self.head.ref # creating a temporary variable. so that temp will point at next node of head node..means starting node+1 .
        prev = self.head # creating a prev variable so that prev is pointing to the head node . 
        for i in range(1,index-1): # looping through linked list until the given index .
            temp = temp.ref # this is for increment for the temporary variable .
            prev = prev.ref # this is for increment for the prev variable . 
        prev.ref = temp.ref # changing the ref of the deleting node and transferring the reference to the before node and connecting the after node.
    
    
    def display(self):
        if self.head is None: # checking the that self.head is null .
            print("the linked list is empty .")
        else:
            length = 0 
            tem = self.head # creating temporary varible self.head
            while tem: # fails until the last node is "NULL"
                length += 1
                print(" --> ",tem.data,end = ' ') # printing the data's 
                tem = tem.ref # increment for while to traverse all the linked nodes.
         
        
        
obj = SLL()
obj.node_begin([1,3,45])
obj.node_begin([2,3])
obj.node_end([3,4])
obj.node_middle(2,[4,345])
obj.node_middle(4,[5,"akil raja"])
obj.display()
# obj.del_beg()
obj.del_end()
#obj.display()
obj.del_pos(1)
obj.display()
```

```python
                      ##  reversing the linked list using normal method ... #3

class node:
    def __init__(self,val):
        self.val = val
        self.ref = None

class SSL:
    def __init__(self):
        self.head = None
    def reverse_ssl(self,nodes):
        # checking if the head Node is Null or not,if Null return None.
        if self.head is None:
            # if the condition is true is None . 
            return None
        # creating a variable assigned as None .
        previous = None
        # creating another node has current which has the pointer and that is pointing to the head
        current = nodes.head 
        # creating another node after which has to iterate towards the end in the linked list .
        after = current.ref
        # the while condition fails until the current is Null . 
        while current :
            # current next is previous means pointing to the Null value at first .
            current.ref = previous
            # changing the previous value has it is iterating the linked list . 
            previous = current
            # making the after has the current 
            current = after 
            # unitl the after became False it will be running .
            if after:
                # incrementing the value of the after 
                after = after.ref
        # making the head to move ahead and storing it in the previous .        
        nodes.head = previous
    
    
    
    def dis(self):
        temp = self.head 
        while temp:
            print(" -->> ",temp.val,end='')
            temp = temp.ref
            
            
n = SSL()  
n.head = node(1)  
n1 = node(2)  
n.head.ref = n1  
n2 = node([1,2,45])  
n1.ref = n2  
n3 = node(4) 
n2.ref = n3 
print("the original linked list is : ")
n.dis()
n.reverse_ssl(n)
print("\n")
print("the reverse linked list is : ")
n.dis()
```

```python
                 ## reversing a linked list using recursive method .. ##

class node:
    def __init__(self,val):
        self.val = val
        self.ref = None

class SSL:
    def __init__(self):
        self.head = None
        
    def sol(self,data):
        new_node = node(data)
        if self.head != None:
            current = self.head 
            while current.next :
                current = current.ref 
            current.ref = new_node
        else:
            return None 
    
    def recusive_reverse(self, current, previous):  

            # If last node mark it head  
            if current.ref is None:  
                self.head = current  

                # Update next to prev node  
                current.ref = previous  
                return  

            # Save curr.next node for recursive call  
            after = current.ref 

            # And update next  
            current.ref = previous  

            self.recusive_reverse(after, current)  

    def reverse(self):  
            if self.head is None:  
                return  
            self.recusive_reverse(self.head, None) 
            
    def dis(self):
        temp = self.head 
        while temp:
            print(" -->> ",temp.val,end='')
            temp = temp.ref
            
            
n = SSL()  
n.head = node(1)  
n1 = node(2)  
n.head.ref = n1  
n2 = node(3)  
n1.ref = n2  
n3 = node(4) 
n2.ref = n3 
print("the original linked list is : ")
n.dis()
n.recusive_reverse(n.head,None)
print("\n")
print("the reverse linked list is : ")
n.dis()    
        
```

```python
class node:
    def __init__(self,data):
        self.data = data
        self.ref = None

class SLL:
    def __init__(self):
        self.head = None
    
    def insert_beg(self,value):
        new_value = node(value)
        new_value.ref = self.head
        self.head = new_value
        
    def insert_end(self,data):
        new_value = node(data)
        temp = self.head
        while temp.ref:
            temp = temp.ref
        temp.ref = new_value
        
    def dis(self):
        temp = self.head
        while temp :
            print(" ->> ",temp.data,end='')
            temp = temp.ref

ob = SLL()
ob.insert_beg([12,34,56])
ob.insert_beg([3,4,56])
ob.insert_end([12,4,5,6])
ob.insert_end([12,3,5,6])
print("linked list is : ")
ob.dis()
```

```python
# Now let's find the top 10 most frequent categories in the X2 column of the dataset.

file.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# making list with top 10 most frequent categories of the variable.

top_10 = [x for x in file.X2.value_counts().sort_values(ascending = False).head(10).index]
top_10
```

```python
# labeling the following top 10 most frequent categories with "1" and "0" or making them into binary values
# here if the variable present in the top 10 will be numbered as "1" otherwise "0" .

for label in top_10:
    file[label] = np.where(file['X2'] == label ,1 ,0)
file[['X2']+top_10].head(40)
```

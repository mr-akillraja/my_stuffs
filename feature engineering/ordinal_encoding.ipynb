# Ordinal Encoding 
 ## - This means arranging the labels in the order/rank and will get the final dataset after encoded.

```python
import pandas as pd
import datetime
```

```python
# making a list of time and date with current date - 20 days before.
df_base = datetime.datetime.today()
df_data_list = [df_base - datetime.timedelta(days=x) for x in range(20)]
```

```python
# making the list into the dataset.
df = pd.DataFrame(df_data_list)
df.columns =['DAY']
df
```

```python
#getting the names of the week.
df['day_week'] = df['DAY'].dt.day_name()
df
```

```python
# sort the labels in the order/rank of the catogory week.
week_map = {'Monday':1,
            'Tuesday':2,
            'Wednesday':3,
            'Thursday':4,
            'Friday':5,
            'Saturday':6,
            'Sunday':7}
df['day_of_week'] = df['day_week'].apply(lambda x: week_map[x])
df.sort_values(by=['day_of_week'])
```

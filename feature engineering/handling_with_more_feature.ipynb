```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib  inline
```

```python
df = pd.read_csv("loan.csv",usecols=['BsmtQual','FireplaceQu','SalePrice','GarageType',])
df.head(20)
```

```python
def impute_nan(df,variable):
    df[variable] = np.where(df[variable].isnull(),"Missing",df[variable])
for i in ['BsmtQual','FireplaceQu','SalePrice','GarageType']:
    impute_nan(df,i)
```

```python
df = pd.read_csv('train.csv',usecols = ['Age','Survived','Fare'])
df.head()
```

```python
df['Age'].dropna().sample(df['Age'].isnull().sum(),random_state=1)
```

```python
def impute_nan(df,variable):
    df[variable+'_random'] = df[variable]
    random_sample = df[variable].dropna().sample(df[variable].isnull().sum(),random_state=0)
    random_sample.index = df[df[variable].isnull()].index
    df.loc[df[variable].isnull(),variable+'_random'] = random_sample
impute_nan(df,'Age')
```
